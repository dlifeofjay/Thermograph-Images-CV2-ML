```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

## DATA LOADING

```python
# Control Group
base_path = r"C:\Users\USER\Documents\ThermoDataBase\Control_Group"
control_group = os.listdir(base_path)

# list to collect results
results = []

for folder in control_group:
    folder_path = os.path.join(base_path, folder)
    
    # Skip if not a folder
    if not os.path.isdir(folder_path):
        continue

    # List all CSVs in the folder (outside Angiosoms)
    lr = os.listdir(folder_path)
    csv_files = [f for f in lr if f.endswith(".csv")]

    # Check for exactly 2 CSVs
    if len(csv_files) != 2:
        print(f"The CSV files in {folder} are not 2")
        continue

    # Read and calculate mean for left and right foot
    df_left = pd.read_csv(os.path.join(folder_path, csv_files[0]), header=None)
    mean_left = df_left[df_left > 0].stack().mean()

    df_right = pd.read_csv(os.path.join(folder_path, csv_files[1]), header=None)
    mean_right = df_right[df_right > 0].stack().mean()

    # Now check for Angiosoms folder
    angio_path = os.path.join(folder_path, "Angiosoms")
    if os.path.exists(angio_path) and os.path.isdir(angio_path):
        angio_files = [f for f in os.listdir(angio_path) if f.endswith(".csv")]

        if len(angio_files) != 8:
            print(f"The CSV files in 'Angiosoms' of {folder} are not 8")
            continue

        # Left Angiosoms
        df_lLCA = pd.read_csv(os.path.join(angio_path, angio_files[0]), header=None)
        mean_lLCA = df_lLCA[df_lLCA > 0].stack().mean()

        df_lLPA = pd.read_csv(os.path.join(angio_path, angio_files[1]), header=None)
        mean_lLPA = df_lLPA[df_lLPA > 0].stack().mean()

        df_lMCA = pd.read_csv(os.path.join(angio_path, angio_files[2]), header=None)
        mean_lMCA = df_lMCA[df_lMCA > 0].stack().mean()

        df_lMPA = pd.read_csv(os.path.join(angio_path, angio_files[3]), header=None)
        mean_lMPA = df_lMPA[df_lMPA > 0].stack().mean()

        # Right Angiosoms
        df_rLCA = pd.read_csv(os.path.join(angio_path, angio_files[4]), header=None)
        mean_rLCA = df_rLCA[df_rLCA > 0].stack().mean()

        df_rLPA = pd.read_csv(os.path.join(angio_path, angio_files[5]), header=None)
        mean_rLPA = df_rLPA[df_rLPA > 0].stack().mean()

        df_rMCA = pd.read_csv(os.path.join(angio_path, angio_files[6]), header=None)
        mean_rMCA = df_rMCA[df_rMCA > 0].stack().mean()

        df_rMPA = pd.read_csv(os.path.join(angio_path, angio_files[7]), header=None)
        mean_rMPA = df_rMPA[df_rMPA > 0].stack().mean()

        
        patient_result = {
            "patient_id": folder,
            "left_foot": mean_left,
            "right_foot": mean_right,
            "left_LCA": mean_lLCA,
            "left_LPA": mean_lLPA,
            "left_MCA": mean_lMCA,
            "left_MPA": mean_lMPA,
            "right_LCA": mean_rLCA,
            "right_LPA": mean_rLPA,
            "right_MCA": mean_rMCA,
            "right_MPA": mean_rMPA
        }

        results.append(patient_result)

# Convert to DataFrame
control_df = pd.DataFrame(results)
```

```python
# DM Group
base_path = r"C:\Users\USER\Documents\ThermoDataBase\DM_Group"
dm_group = os.listdir(base_path)

dm_result = []

for folder in dm_group:
    folder_dir = os.path.join(base_path, folder)

    if not os.path.isdir(folder_dir):
        continue

    lr = os.listdir(folder_dir)

    csv_files = [f for f in lr if f.endswith(".csv")]

    if len(csv_files) != 2:
        print(f"The csv files in {folder} is not 2")
        continue

    #Reading and estimating mean of csv
    df_left = pd.read_csv(os.path.join(folder_dir, csv_files[0]), header=None)
    mean_left = df_left[df_left > 0].stack().mean()

    df_right = pd.read_csv(os.path.join(folder_dir, csv_files[1]), header=None)
    mean_right = df_right[df_right > 0].stack().mean()

    # Now for the Angiosoms folder

    angio_path = os.path.join(folder_dir, "Angiosoms")
    if os.path.exists(angio_path) and os.path.isdir(angio_path):
        angio_files = [f for f in os.listdir(angio_path) if f.endswith(".csv")]

        if len(angio_files) != 8:
            print(f"The csv in {angio_path} is not 8")
            continue

        # Left Angiosoms
        df_lLCA = pd.read_csv(os.path.join(angio_path, angio_files[0]), header=None)
        mean_lLCA = df_lLCA[df_lLCA > 0].stack().mean()

        df_lLPA = pd.read_csv(os.path.join(angio_path, angio_files[1]), header=None)
        mean_lLPA = df_lLPA[df_lLPA > 0].stack().mean()

        df_lMCA = pd.read_csv(os.path.join(angio_path, angio_files[2]), header=None)
        mean_lMCA = df_lMCA[df_lMCA > 0].stack().mean()

        df_lMPA = pd.read_csv(os.path.join(angio_path, angio_files[3]), header=None)
        mean_lMPA = df_lMPA[df_lMPA > 0].stack().mean()

        # Right Angiosoms
        df_rLCA = pd.read_csv(os.path.join(angio_path, angio_files[4]), header=None)
        mean_rLCA = df_rLCA[df_rLCA > 0].stack().mean()

        df_rLPA = pd.read_csv(os.path.join(angio_path, angio_files[5]), header=None)
        mean_rLPA = df_rLPA[df_rLPA > 0].stack().mean()

        df_rMCA = pd.read_csv(os.path.join(angio_path, angio_files[6]), header=None)
        mean_rMCA = df_rMCA[df_rMCA > 0].stack().mean()

        df_rMPA = pd.read_csv(os.path.join(angio_path, angio_files[7]), header=None)
        mean_rMPA = df_rMPA[df_rMPA > 0].stack().mean()

        
        patient_result = {
            "patient_id": folder,
            "left_foot": mean_left,
            "right_foot": mean_right,
            "left_LCA": mean_lLCA,
            "left_LPA": mean_lLPA,
            "left_MCA": mean_lMCA,
            "left_MPA": mean_lMPA,
            "right_LCA": mean_rLCA,
            "right_LPA": mean_rLPA,
            "right_MCA": mean_rMCA,
            "right_MPA": mean_rMPA
        }

        dm_result.append(patient_result)

# Convert to DataFrame
DM_df = pd.DataFrame(dm_result)
```

### SYMMETRY ANALYSIS(LEFT FOOT - RIGHT FOOT)

```python
control_df.columns
```

```python
def lr(data):
    data["lr_LCA"] = np.abs(data["left_LCA"] - data["right_LCA"])
    data["lr_LPA"] = np.abs(data["left_LPA"] - data["right_LPA"])
    data["lr_MCA"] = np.abs(data["left_MCA"] - data["right_MCA"])
    data["lr_MPA"] = np.abs(data["left_MPA"] - data["right_MPA"])
    
    return data
```

```python
control_df = lr(control_df)
```

```python
#CONTROL GROUP
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Thermal Symmetry Distribution (T) across Angiosomes (CONTROL GROUP)', fontsize=16)

features = ["lr_LCA", "lr_LPA", "lr_MCA", "lr_MPA"]

for i, col in enumerate(features):
    ax = axes[i//2, i%2]
    
    sns.histplot(control_df[col], kde=True, ax=ax, color='skyblue', bins=15)
    
    mean_val = control_df[col].mean()
    ax.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
    
    ax.axvline(2.2, color='darkorange', linestyle='-', label='Threshold (2.2C)')
    
    ax.set_title(f'Control Group Symmetry Distribution: {col}')
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("control_symmetry.png")
plt.show()
```

## THERMAL CHANGE INDEX VALIDATION

```python
df_cols = ['left_LCA', 'left_LPA',
   'left_MCA', 'left_MPA', 'right_LCA', 'right_LPA', 'right_MCA',
   'right_MPA']
control_means = control_df[df_cols].mean().values
def calculate_tci(data, control_means):
    data["left_TCI"] = (
        (data["left_LCA"] - control_means[0]) + 
        (data["left_LPA"] - control_means[1]) + 
        (data["left_MCA"] - control_means[2]) + 
        (data["left_MPA"] - control_means[3])
    ) / 4

    data["right_TCI"] = (
        (data["right_LCA"] - control_means[4]) + 
        (data["right_LPA"] - control_means[5]) + 
        (data["right_MCA"] - control_means[6]) + 
        (data["right_MPA"] - control_means[7])
    ) / 4

    return data
```

```python
control_df = calculate_tci(control_df, control_means)
```

```python
# CONTROL GROUP
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('Global Thermal Change Index (TCI) Audit: Left vs. Right Foot (CONTROL GROUP)', fontsize=16)

tci_cols = ["left_TCI", "right_TCI"]
titles = ["Left Foot TCI", "Right Foot TCI"]

for i, col in enumerate(tci_cols):
    ax = axes[i]
    
    sns.histplot(data=control_df, x=col, kde=True, ax=ax, alpha=0.6)
    
    ax.axvline(0, color='black', linestyle='--', label='Healthy Baseline (0.0)')
    
    ax.axvline(2.0, color='red', linestyle=':', label='Possible Noise Threshold (2.0)')
    
    ax.set_title(titles[i])
    ax.set_xlabel("TCI Value")
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("control_tci.png")
plt.show()
```

```python
DM_df = calculate_tci(DM_df, control_means)
```

```python
#DM GROUP
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('Global Thermal Change Index (TCI) Audit: Left vs. Right Foot (DM GROUP)', fontsize=16)

tci_cols = ["left_TCI", "right_TCI"]
titles = ["Left Foot TCI", "Right Foot TCI"]

for i, col in enumerate(tci_cols):
    ax = axes[i]
    sns.histplot(data= DM_df, x=col, kde=True, ax=ax, alpha=0.6)

    ax.axvline(0, color='black', linestyle='--', label = 'Healthy Baseline (0.0)')

    ax.axvline(2.0, color='red', linestyle=':', label = 'Clinical Risk Threshold (2.0)')

    ax.set_title(titles[i])
    ax.set_xlabel("TCI Value")
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("DM_tci.png")
plt.show
```
